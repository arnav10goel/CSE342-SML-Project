```python
# Split the X_train and y_train into training and validation sets
X_train2, X_val, y_train2, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=0)
# Applying Logistic Regression

from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(random_state = 0, max_iter=1000)

classifier.fit(X_train2, y_train2)

# Predicting the Test set results
y_pred = classifier.predict(X_val)

# Accuracy of the model
from sklearn.metrics import accuracy_score

print(accuracy_score(y_val, y_pred))
```

```python
# Create a random forest classifier
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=1000, bootstrap=False, max_depth=90, min_samples_leaf=3, min_samples_split=6, n_jobs=-1, random_state=42, verbose=1)

# Train the classifier
rf.fit(X_train2, y_train2)

# Predict the validation set
y_pred = rf.predict(X_val)

# Create a submission file
# Opening the file to write the predictions
# file = open('RFC.csv', 'w')

# # Writing the header
# file.write('ID,Category')
# for i in range(len(y_pred)):
#     file.write('\n')
#     file.write(str(i))
#     file.write(',')
#     file.write(y_pred[i])

# file.close()
```

```python
# Accuracy of the model
from sklearn.metrics import accuracy_score

print(accuracy_score(y_val, y_pred))
```

```python
# Import a Bagging Classifier
from sklearn.ensemble import BaggingClassifier

# Create a bagging classifier
bg = BaggingClassifier(rf, max_samples=0.5, max_features=0.5, n_estimators=1000, verbose=1)

# Train the classifier
bg.fit(X_train2, y_train2)

# Predict the validation set
y_pred = bg.predict(X_val)

# Accuracy of the model
print(accuracy_score(y_val, y_pred))
```

```python
# Predict the test set
y_pred = rf.predict(X_test)

# Create a submission file
# Opening the file to write the predictions
file = open('PCA_LDA_Clustering.csv', 'w')

# Writing the header
file.write('ID,Category')
for i in range(len(y_pred)):
    file.write('\n')
    file.write(str(i))
    file.write(',')
    file.write(y_pred[i])

file.close()
```
