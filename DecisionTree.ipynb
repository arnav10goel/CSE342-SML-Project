```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data_matrix[:, :-1], data_matrix[:, -1], test_size=0.2, random_state=42)

# Create a logistic regression classifier
clf = LogisticRegression(max_iter=1000, random_state=0, solver='newton-cg', multi_class='multinomial')

# Train the classifier
clf.fit(X_train, y_train)

# Predict the labels of the test set
y_pred = clf.predict(X_test)

# Print the accuracy
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import BaggingClassifier

# Create a random forest classifier
rfc = RandomForestClassifier(bootstrap=True, max_depth=70, max_features='sqrt', min_samples_leaf=4, min_samples_split=10, n_estimators=400)

# Train the classifier
rfc.fit(X_train, y_train)

# Predict the labels of the test set
y_pred = rfc.predict(X_test)

# Print the accuracy
print(accuracy_score(y_test, y_pred))

# Create a bagging classifier
bg = BaggingClassifier(rfc, max_samples=0.5, max_features=0.5, n_estimators=400)

# Train the classifier
bg.fit(X_train, y_train)

# Predict the labels of the test set
y_pred = bg.predict(X_test)

# Print the accuracy
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.ensemble import AdaBoostClassifier

# Make the AdaBoost classifier
clf_boost = AdaBoostClassifier(rfc, n_estimators = 10)

clf_boost.fit(X_train, y_train)

y_pred = clf_boost.predict(X_test)

print(accuracy_score(y_test, y_pred))
```
